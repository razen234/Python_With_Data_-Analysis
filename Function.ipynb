# FUNCTION
### Function: 
1. a block of code run only when it is called
2. function can return data as a result
3. avoid helping code repeat
# Syntax:
##  def my_function()
### print("Hello from a function ")


```python
def my_function():
    print("Hello From a Function")

my_function()
```

```python
def my_function():
    print("Hello From a Function")

for i in range(5):
    my_function()
```

```python
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5.0/9.0
    return celsius

temp_f = float(input("Enter the temperature in Fahrenheit: "))
temp_c = fahrenheit_to_celsius(temp_f)
print(f"{temp_f} degrees Fahrenheit is equal to {temp_c} degrees Celsius.")
```
