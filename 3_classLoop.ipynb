```python
#Prime number 
x = int(input("Enter the number "))
if x <= 1:
    print("Not a prime number")
else:
    for i in range(2, x):
        if (x % i) == 0:
            print("Not a prime number")
    else:
        print(" It's a prime number")
    
```

## WAP to find the number is a palindrome or not: 121, 424, 323, 123321

```python
import math
number = int(input("Enter the number to check if it is palindrome or not: "))
rev = 0
temp = number 
if number > 0 : 
    digit = int(math.log10(number)) + 1
else:
    digit = 1
    
for _ in range(digit):
    last_digit = temp % 10
    rev = (rev * 10) + last_digit
    temp //= 10

if number == rev:
    print(number , "It is palindrome ")
else:
    print(number ," Is not palindrome")
```
