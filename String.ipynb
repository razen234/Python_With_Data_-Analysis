## String Immutability
### Strings are immutable, which means that they cannot be changed after they are created. If we need to manipulate strings then we can use methods like concatenation, slicing or formatting to create new strings based on original.
### Example: In this example we are changing first character by building a new string.

```python
s = "Raj"
s = "S" + s[1:]
print(s)
```

## Deleting a String
### In Python, it is not possible to delete individual characters from a string since strings are immutable. However, we can delete an entire string variable using the del keyword.
### Example: Here, we are using del keyword to delete a string.

```python
a = "Simran"
del a 
print(a)
### working 
```

## Updating a String
### As strings are immutable, “updates” create new strings using slicing or methods such as replace().
### Example: This code fix the first letter and replace a word.

```python
a = "Dharamraj"
a1 = "d" + a[1:] # updating the string 
a2=  a.replace("Dharamraj", "Raj")
print(a1)
print(a2)
```

```python
from datetime import datetime

now = datetime.now()
date_string = now.strftime("%y-%m-%d %H:%M")

bill_lines = ""
sub_total = 0

n = int(input("Enter n = "))
for i in range(n):
    name = input("Product Name: ")
    price = int(input("Entert price = "))
    qty = float(input("Enter qty: "))
    total = price * qty

    info = f"{name:<15} {price:<10} {qty:<5} {total:<10.2f}\n"
    bill_lines = bill_lines + info
    sub_total = sub_total + total

vat_rate = 0.13
vat_amount = sub_total * vat_rate
grand_total = sub_total + vat_amount

print("\n" + "="*45)
print("      SAMIRAJ KIRANA PASAL BHAKTAPUR")
print(f"Date: {date_string}")
print("="*45)


print(f"{'Product':<15} {'Price':<10} {'Qty':<5} {'Total':<10}")
print("-" * 45)


print(bill_lines, end="")

print("-" * 45)
# Totals and Tax Breakdown
print(f"{'Sub-Total':<32} Rs. {sub_total:>8.2f}")
print(f"{'VAT (13%)':<32} Rs. {vat_amount:>8.2f}")
print("-" * 45)
print(f"{'GRAND TOTAL':<32} Rs. {grand_total:>8.2f}")
print("="*45)
print("        Thank you for your visit!")
```

```python
s = int(input("Enter the total number of students: "))
database = {}
for i in range(s):
    print(f"\n--- Entering data for Student {i+1} ---")
    roll = input("Enter Roll Number: ")
    name = input("Enter Student Name: ")
    phy = int(input("Enter mark of PHY: "))
    che = int(input("Enter mark of CHE: "))
    nep = int(input("Enter mark of NEP: "))
    mat = int(input("Enter mark of MAT: "))
    eng = int(input("Enter mark of ENG: "))
    com = int(input("Enter mark of COM: "))
    total_mark = phy + che + nep + mat + eng + com 
    average = total_mark / 6
    cgpa = average / 25 

    database[roll] = {
        "name": name,
        "marks":{"PHY":phy, "CHE":che, "NEP":nep, "ENG":eng, "MAT": mat, "COM":com},
        "total":total_mark,
        "avg": round(average, 2),
        "cgpa": round(cgpa, 2)
    }

while True:
    print("\n" + "="*30)
    search_roll = input("Enter Roll Number to view Marksheet (or 'exit' to quit): ")
    
    if search_roll.lower() == 'exit':
        break
    
    if search_roll in database:
        std = database[search_roll]
        print("\n" + "*"*40)
        print(f"       STUDENT MARKSHEET")
        print("*"*40)
        print(f"Name: {std['name']}")
        print(f"Roll No: {search_roll}")
        print("-" * 40)
        
        # Loop through marks to display subjects
        for subject, score in std['marks'].items():
            print(f"{subject:<10}: {score}")
            
        print("-" * 40)
        print(f"GRAND TOTAL: {std['total']}/600")
        print(f"AVERAGE    : {std['avg']}%")
        print(f"CGPA       : {std['cgpa']}")
        print("*"*40)
    else:
        print("Error: Roll Number not found!")
```
